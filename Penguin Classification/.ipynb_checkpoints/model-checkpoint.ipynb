```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
import pickle
```

```python
data = pd.read_csv('penguins_size.csv')
data.head()
```

```python
data.isna().sum()
```

```python
data.shape
```

```python
target = 'sex'
encode = ['species', 'island']
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data, dummy], axis=1)
    del data[col]
    
data.head()
```

```python
data = data.dropna()
data.shape
```

```python
data
```

```python
data.to_csv('penguins_cleaned')
```

```python
target_wrapper = {'MALE':0, 'FEMALE':1}
def target_encode(val):
    return target_wrapper[val]

data['sex'] = data['sex'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('sex', axis=1)
Y = data['sex']
```

```python
clf = RandomForestClassifier()
clf.fit(X, Y)
```

```python
pickle.dump(clf, open('penguins_clf.pkl', 'wb'))
```
